```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))
```

```python
lamb_og, inten_og=np.loadtxt(r'C:\Users\jojos\Himmel auf.txt', skiprows=17, 
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
lamb_mg, inten_mg=np.loadtxt(r'C:\Users\jojos\himmel zu.txt', skiprows=17, 
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
lamb_s, inten_s=np.loadtxt(r'C:\Users\jojos\Sonne.txt', skiprows=17, 
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
```

```python
plt.plot(lamb_og, inten_og, label='ohne Fenster')
plt.plot(lamb_mg, inten_mg, label='mit Fenster')
plt.plot(lamb_s, inten_s, label='Sonne')
plt.title('Gemessenes Sonnespektrum direkt in Sonne, sowie mit und ohne Fenster')
plt.xlabel('Wellenlänge \nm')
plt.ylabel('Intensitaet \ b.E.')
plt.legend()
plt.grid()
plt.ylim((0,60000))
plt.xlim((250,900))
plt.show()
```

```python
A=1-inten_mg/inten_og
plt.plot(lamb_mg,A)
plt.title('Absorption von Glas')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Absorption / b.E.')
plt.ylim((0,1))
plt.xlim((320,800))
plt.show()
```

```python
plt.plot(lamb_og,inten_og, label='ohne Fenster')
plt.plot(lamb_s,inten_s,label='Sonnenlicht')
plt.title('Sonnenlicht und Streulicht')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Intensität / b.E.')
plt.legend()
plt.xlim((320,800))
plt.show()
```

```python
B=inten_og/(inten_s+1)
plt.plot(lamb_mg,B)
plt.title('Verschiebung')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Absorption / b.E.')
plt.xlim((320,800))
plt.ylim((0,10))
plt.show()
```
